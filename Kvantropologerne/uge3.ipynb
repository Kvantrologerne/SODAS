```python
import pandas as pd
import re
from wordcloud import WordCloud
```

```python
uffe = pd.read_csv('data/uffeelbaek.csv')
```

```python
alltext = ""
for tweet in uffe.text:
    tweet = tweet.split("https")[0]
    alltext = alltext + " " + tweet
```

```python
allwords = re.split(r'\W+', alltext)
```
